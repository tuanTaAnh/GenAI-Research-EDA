```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.miscmodels.ordinal_model import OrderedModel
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import kruskal
```

```python
df_whole = pd.read_csv(r'Data/Higher Education Students Early Perceptions of ChatGPT Global Survey Data/final dataset.csv')
df_whole
```

```python
columns = ['citizenship','gender','study_country','institution_public','student_status','study_level','bachelor_first_year','field_of_study','mode_current_study']
df = df_whole.groupby(columns,group_keys = False).apply(lambda x: x.sample(frac = 0.7, random_state = 42))
print(len(df_whole), len(df))
df
```

```python
df = df.drop(['citizenship','institution_name','study_country','general_views_chatgpt','email_contact'],axis = 1)
```

```python
for col in df:
    df[col] = df[col].fillna(df[col].median())
```

```python
mapping = {1:1,2:1,3:2,4:3,5:3}
df['use_extent'] = df['use_extent'].map(mapping)
```

### Research Question 1: Are students more satisfied with Generative AI than traditional forms of learning?

```python
mapping = {1:1,2:1,3:2,4:3,5:3}
gen_over_trad = ["sat_vs_google","sat_vs_professor_interaction","sat_info_professors","sat_info_accuracy"]
for i in gen_over_trad:
     df[i] = df[i].map(mapping)
```

```python
for col in gen_over_trad:
    df[col] = df[col].fillna(df[col].median())
```

```python
for i in gen_over_trad:
    model = OrderedModel(df[i],df['use_extent'],distr = 'logit')
    res = model.fit(disp = False)
    print(i,'p_value:', res.pvalues['use_extent'], 'Odds Ratio', np.exp(res.params['use_extent']))
```

### Research Question 1: Do students prefer generative AI over traditional methods of learning?

```python
easier_than_professors_comm = df.groupby(['use_extent','sat_vs_professor_interaction']).size().reset_index(name = 'count')
easier_than_professors_comm['percentage'] = easier_than_professors_comm.groupby('use_extent')['count'].transform(lambda x: x/x.sum()*100)
easier_than_professors_comm['percentage'] = easier_than_professors_comm['percentage'].round()
easier_than_professors_comm
```
